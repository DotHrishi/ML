```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans, SpectralClustering
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

df=pd.read_csv("drug200.csv")
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.dtypes
```

```python
df=pd.get_dummies(df,columns=['Sex','BP','Cholesterol'],drop_first=True)
```

```python
df.head()
```

```python
df=df.drop(columns=['Na_to_K'])
```

```python
X=df.drop(columns=['Drug'])
```

```python
km=KMeans(n_clusters=5)
ss=StandardScaler()
X=ss.fit_transform(X)
```

```python
df['cluster']=km.fit_predict(X)
```

```python
df.head()
```

```python
print(silhouette_score(X,df['cluster']))
```

```python
db=DBSCAN(eps=1.5,min_samples=5)
df['cluster_db']=db.fit_predict(X)

print(silhouette_score(X,df['cluster_db']))
```

```python
sc=SpectralClustering()
df['cluster_spec']=sc.fit_predict(X)

print(silhouette_score(X,df['cluster_spec']))
```
