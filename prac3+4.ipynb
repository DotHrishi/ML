```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

df=pd.read_csv("drug200.csv")

df.head()
df.info()
df.describe()
df.dtypes
```

```python
df.head()
df.tail()
```

```python
print(df.isnull().sum())
```

```python
df=pd.get_dummies(df,columns=["Age","Sex","BP","Cholesterol"])
```

```python
df.drop(columns=["Na_to_K"])
```

```python
X=df.drop(columns=["Drug"])
y=df["Drug"]
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.transform(X_test)
```

```python
dt=DecisionTreeClassifier()
dt.fit(X,y)
```

```python
y_pred=dt.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
print(classification_report(y_test,y_pred))
print(cm)
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(X_train,y_train)
y_pred2=svc.predict(X_test)

print(classification_report(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
```

```python
kfol=KFold(random_state=42,shuffle=True,n_splits=5)
cv_score=cross_val_score(DecisionTreeClassifier(),X,y,cv=kfol,scoring="accuracy")
print(cv_score)
```
